```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np
import pandas as pd

sns.set(style="darkgrid")
```

# Mengambil directory dan input file awal

```python
cwd = os.getcwd()
input_dir = os.path.join(cwd, 'input')
output_dir = os.path.join(cwd, 'output')
features = os.path.join(output_dir, 'relevance_features.csv')
```

```python
df = pd.read_csv(features)
```

```python
df
```

---

### Menghitung nilai korelasi antar features

```python
correlated_value = df.corr()
```

```python
correlated_value
```

Membatasi nilai minimal korelasi

```python
threshold = 0.9
```

Mengambil seluruh nilai yang berada di atas *threshold*

```python
high_corrs = (correlated_value[abs(correlated_value) > threshold][correlated_value != 1.0]).unstack().dropna().to_dict()
```

Mengambil nilai pasangan *feature* (tanpa duplikat) yang memiliki korelasi di atas *threshold*. Serta melakukan *sorting* dari nilai korelasi yang memiliki nilai tertinggi

```python
unique_high_corrs = pd.DataFrame(list(set([(tuple(sorted(key)), high_corrs[key]) for key in high_corrs])), columns=['feature_pair', 'correlation_coefficient'])
unique_high_corrs = unique_high_corrs.loc[abs(unique_high_corrs['correlation_coefficient']).argsort()[::-1]]
```

Melihat nilai pasangan *features* dan nilai korelasinya

```python
unique_high_corrs
```

```python
unique_high_corrs.to_csv(os.path.join(output_dir, 'high_pair_correlated_features.csv'), index=False)
```

---

## Mengambil nilai per *features* untuk kemudian dilakukan penghapusan *features* pada *dataframe*

```python
correlated_features = set()

for uniques in unique_high_corrs['feature_pair']:
    for unique in uniques:
        correlated_features.add(unique)
```

```python
len(correlated_features)
```

Menyimpan *features* yang memiliki korelasi tinggi

```python
corr = pd.DataFrame(list(correlated_features))
corr.to_csv(os.path.join(output_dir, 'high_correlated_features.csv'), index=False)
```

```python
df = df.drop(list(correlated_features), axis=1)
```

```python
df
```

Menyimpan *fetures* tanpa *features* yang berkorelasi tinggi. Features ini akan digunakan untuk dilakukan training model menggunakan Random Forest Classifier

```python
df.to_csv(os.path.join(output_dir, 'features_without_high_correlated_value.csv'), index=False)
```
